```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_real = pd.read_csv('./data/btc.csv')
df_fake = pd.read_csv('./predict/BOOSTING/final_pred_boosting_gru_gradient.csv')
```

```python
real_price = df_real['close'].values.reshape(-1,1)
real_price = np.array(real_price)
real_price = real_price.reshape(real_price.shape[0], 1)

predict_price = df_fake['close'].values.reshape(-1,1)
predict_price = np.array(predict_price)
predict_price = predict_price.reshape(predict_price.shape[0], 1)


plt.figure(figsize=(70,10))
plt.plot(real_price, color="red", label=f"Real BTC Closing Prices")
plt.plot(predict_price, color="blue", label=f"Predicted BTC Closing Prices")
plt.title(f"BTC Prices")
plt.xlabel("Time (day)")
plt.ylabel("Stock Prices")
plt.ylim(bottom=0)
plt.legend()
plt.show()
```
